## 6. Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function


```python
def cleanse_year(year_str):  # defines function to cleanse year data
    if "*" in year_str:
        year = year_str.replace("*", "")  # removes asterisk from year data if present
    else:
        year = year_str  # if asterisk not present, data point remains the same
    
    return int(year)  # converts all year data to integers

df["year_cleansed"] = df["Census Year"].apply(cleanse_year)  # adds "year_cleansed" column to data frame

df  # prints data frame
```

```python
def cleanse_pop(pop_str):  # defines function to cleanse population data
    if "," in pop_str:
        pop = pop_str.replace(",", "")  # removes comma from population data if present
    else:
        pop = pop_str  # if comma not present, data point remains the same
    
    return int(pop)  # converts all population data to integers

df["population_cleansed"] = df["Total Population"].apply(cleanse_pop)  # adds "population_cleansed" column to data frame

df  # prints data frame
```

```python
def cleanse_cost(cost_str):  # defines function to cleanse cost data
    if "$" in cost_str:
        cost_comma = cost_str.replace("$", "")  # removes dollar sign from cost data if present
        cost_billion = cost_comma.replace(",", "")  # removes comma from cost data if present
        cost_decimal = cost_billion.replace(" Billion", "000000000")  # replaces " Billion" from cost data with nine 0s if present
        cost = cost_decimal.replace(".5000000000", "500000000")  # replaces decimal point and removes a 0 from "4.5 billion" cost data point to give correct value
    else:
        cost = cost_str  # if dollar signs not present, data point remains the same
    
    return int(cost)  # converts all cost data to integers

df["cost_cleansed"] = df["Census Cost"].apply(cleanse_cost)  # adds "cost_cleansed" column to data frame

df  # prints data frame
```

```python
def cleanse_avg(avg_str):  # defines function to cleanse average cost data
    if "$" in avg_str:
        avg = avg_str.replace("$", "")  # removes dollar sign from average cost data if present
    elif " cents" in avg_str:
        avg_cents = avg_str.replace(" cents", "")  # removes " cents" from average cost data if present
        avg = float(avg_cents)/100  # converts data points that contained " cents" to dollar values by dividing by 100
    else:
        avg = avg_str  # if dollar signs or " cents" not present, data point remains the same
    
    return float(avg)  # converts all average cost data to floating point values

df["average_cleansed"] = df["Average Cost Per Person"].apply(cleanse_avg)  # adds "average_cleansed" column to data frame

df  # prints data frame
```
