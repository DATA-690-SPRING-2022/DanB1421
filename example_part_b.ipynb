```python
import pandas as pd
import plotly
import plotly.express as px

DATA_URL = "https://raw.githubusercontent.com/DanB1421/DanB1421/main/wdi_data_3_28.csv"
```

```python
df = pd.read_csv(DATA_URL)

df.sample(10)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
df_pivot = df.pivot_table(
    index=["Year", "Country Code", "Country Name", "Region", "Income Group",	"Lending Type"], 
    columns="indicator", 
    values="value"
)

df_pivot.head()
```

```python
df_pivot.reset_index(inplace=True)

df_pivot.head()
```

```python
df_pivot.query("Year == 2018").sample(10)
```

```python
df_scatter = df_pivot[df_pivot["Country Code"].isin(["CHN","IND", "RUS", "USA"])].query("Year == 2018")
fig = px.scatter(
    df_scatter,
    title="Wealth vs Health Based on Year 2018 Data from World Bank",
    x = "NY.GDP.PCAP.PP.CD",
    y = "SP.DYN.LE00.IN",
    size="SP.POP.TOTL",
    text="Country Name",
    color="Region",
    template="plotly_dark",
    labels={"NY.GDP.PCAP.PP.CD":"GDP Per Capita", "SP.DYN.LE00.IN":"Life Expectancy"}
)

fig.update_traces(textposition='top left')
fig.show()
```

```python
df_pivot.head()
```

```python
df_2018 = df_pivot.query("Year == 2018")

df_2018.head()
```

```python
df_2018_groupby = df_2018.groupby(["Income Group"]).agg(
    gdp_mean = ('NY.GDP.PCAP.PP.CD', 'mean'),
    life_mean = ('SP.DYN.LE00.IN', 'mean'),
    pop_sum = ('SP.POP.TOTL', 'sum')
).reset_index()

df_2018_groupby
```

```python
fig = px.scatter(
    df_2018_groupby,
    title="Wealth vs Health Based on Year 2018 Data from World Bank",
    x = "gdp_mean",
    y = "life_mean",
    size = "pop_sum",
    size_max=100,
    text="Income Group",
    color="Income Group",
    template="plotly_dark",
    labels={"gdp_mean":"GDP Per Capita", "life_mean":"Life Expectancy"}
)

fig.update_traces(textposition='top center')
fig.show()
```

```python
df_2020_pop = df.query("Year == 2020").query("indicator == 'SP.POP.TOTL'")
df_2020_pop.sample(5)
```

```python
df_2020_pop_sort = df_2020_pop.sort_values(by="value", ascending=False)

df_2020_pop_sort.head()
```

```python
fig = px.bar(
    df_2020_pop_sort, 
    x="Country Name",
    labels = {"value":"2020 Population", "Country Name":""}, 
    y="value",
    color = "Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5]
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
df_group_sum = df_2020_pop.groupby('Region').sum().reset_index().sort_values(by="value", ascending=False)

df_group_sum
```

```python
fig = px.bar(
    df_group_sum, 
    x="Region",
    labels = {"value":"2020 Population", "Region":""}, 
    y="value",
    color = "Region",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title='Comparing Total Populations on All Continents'
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
df_2020_gdp = df.query("Year == 2020").query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_2020_gdp.sample(5)
```

```python
df_group_mean = df_2020_gdp.groupby('Region').mean().reset_index().sort_values(by="value", ascending=False)

df_group_mean
```

```python
fig = px.bar(
    df_group_mean, 
    x="Region",
    labels = {"value":"2020 GDP Per Capita (group mean)", "Region":""}, 
    y="value",
    color = "Region",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="Comparing Mean GDP Per Capita on All Continents"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
df_gdp = df.query("indicator == 'NY.GDP.PCAP.PP.CD'") 
df_gdp = df_gdp[df_gdp["Country Code"].isin(["CHN","IND", "RUS", "USA"])]
#df_gdp ["Year"] = df_gdp["Year"].astype("str")
df_gdp.info()
```

```python
fig = px.line(
    df_gdp, 
    x="Year",
    labels = {"value": "GDP Per Capita"}, 
    y="value",
    color = "Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)

fig.update_layout(showlegend=True)
fig.update_xaxes(dtick="Y1")   
fig.show()
```

```python
fig = px.line(
    df_gdp, 
    x="Year", 
    labels={"value":"GDP Per Capita"},
    y="value",
    color="Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)
fig.update_xaxes(dtick="Y1")  
fig.show()
```

```python
df_gdp2 = df_gdp

#df_gdp2["Year"] = df_gdp2["Year"].astype(str)


fig = px.line(
    df_gdp2, 
    x="Year", 
    labels={"value":"GDP Per Capita"},
    y="value",
    color="Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)



_df = df_gdp2.query("Year == 2020")
for index, row in _df.iterrows():
    anno_dict = dict(
        font=dict(color="white"),
        x=row["Year"],
        y=row["value"],
#                        opacity=0.50,
        showarrow=False,
        text=row["Country Name"],
        textangle=0,
        xanchor='left',
        yanchor="bottom",
        xref="x",
        yref="y"
    )
    fig.add_annotation(anno_dict)  

fig.update_xaxes(dtick="Y1")       # tick interval very year.
fig.update_layout(showlegend=False)
fig.show()
```
